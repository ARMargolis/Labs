### Derived from client-python/examples/example2.py:

This example shows listening for up to 10 events, within a time limit of 5 seconds:

````python
# Derived from example2:

from kubernetes import watch

#watch.Watch?

count = 10
w = watch.Watch()
for event in w.stream(v1.list_namespace, timeout_seconds=5):
    print("count %s: Event: %s %s" % (count, event['type'], event['object'].metadata.name))
    count -= 1
    if not count:
        w.stop()

print("Ended.")
````

### Derived from client-python/examples/example3.py:

This example shows the available APIs.

So far we have been using the CoreV1Api, but other kubernetes apis are available, some with different versions as listed below:

````python
print("Supported APIs (* is preferred version):")
print("    %-40s %s" %
          ("core", ",".join(client.CoreApi().get_api_versions().versions)))

for api in client.ApisApi().get_api_versions().groups:
    versions = []
    for v in api.versions:
        name = ""
        if v.version == api.preferred_version.version and len(
                api.versions) > 1:
            name += "*"
        name += v.version
        versions.append(name)
    print("    %-40s %s" % (api.name, ",".join(versions)))
````

````python
from kubernetes.client.apis.apps_v1beta1_api import AppsV1beta1Api
from kubernetes.client.apis import extensions_v1beta1_api

def cleanup_pods_for_context(context='default'):
    contexts, active_context = config.list_kube_config_contexts()
    if not contexts:
        print("Cannot find any context in kube-config file.")
        return

    contexts = [context['name'] for context in contexts]
    print("Available contexts:")
    pprint.pprint(contexts)
    
    if not context in contexts:
        print("Error {} is not an available context".format(context))
        return
    
    active_index = contexts.index(active_context['name'])

    # Configs can be set in Configuration class directly or using helper utility
    config.load_kube_config(context=context)
        
    print("Cluster host=" + configuration.host)
    
    v1 = client.CoreV1Api()
    v1beta = client.AppsV1beta1Api()
    v1betaExt = client.ExtensionsV1beta1Api()
    
    # Loop through namespaces, then destroy object types one by one: in what order????    
    print(); print("Namespaces:")
    for ns in v1.list_namespace().items:
        if ns.metadata.name[:4] == "kube":
            print(" - " + ns.metadata.name + " - SKIPPED")
            continue
    
        print(" - " + ns.metadata.name)
        
       # TODO - Try: DeleteOptions.PropagationPolicy="Foreground" to delete dependent objects ...

        # DELETE deployments:
        for deploy in v1beta.list_namespaced_deployment(ns.metadata.name).items:
            print("    Deleting deploy {}/{}".format(ns.metadata.name, deploy.metadata.name))
            v1beta.delete_namespaced_deployment(deploy.metadata.name, ns.metadata.name, body={})
                
        # DELETE replication_controllers:
        #print("replication controllers = " + str( v1.list_namespaced_replication_controller(ns.metadata.name).items ))
        for replication_controller in v1.list_namespaced_replication_controller(ns.metadata.name).items:
            print("    Deleting replication_controller {}/{}".format(ns.metadata.name, replication_controller.metadata.name))
            v1.delete_namespaced_replication_controller(replication_controller.metadata.name, ns.metadata.name, body={})
        
        # DELETE replica_sets:
        #print("replica     sets = "        + str( v1betaExt.list_namespaced_replica_set(ns.metadata.name).items ))
        for replica_set in v1betaExt.list_namespaced_replica_set(ns.metadata.name).items:
            print("    Deleting replica_set {}/{}".format(ns.metadata.name, replica_set.metadata.name))
            v1betaExt.delete_namespaced_replica_set(replica_set.metadata.name, ns.metadata.name, body={})
        
        # DELETE pods:
        for pod in v1.list_namespaced_pod(ns.metadata.name).items:
            #print("   - pod <ip:{}, host:{}>: {}".format(pod.status.pod_ip, pod.status.host_ip, pod.metadata.name))
            #print("   - pod <ip:{}>: {}".format(pod.status.pod_ip, pod.metadata.name))
            print("Deleting pod {}/{}".format(ns.metadata.name, pod.metadata.name))
            v1.delete_namespaced_pod(pod.metadata.name, ns.metadata.name, body={})
            
cleanup_pods_for_context('minikube')
````
